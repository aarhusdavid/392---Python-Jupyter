```python
iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print(iris.head())
```

```python
(ggplot(iris,aes(x = 'sepal_length',
                      y = 'sepal_width',
                color = 'species')) + geom_point())
```

```python
(ggplot(iris, aes('sepal_length')) + geom_histogram())
```

```python
(ggplot(iris, aes('species')) + geom_bar())
```

```python
(ggplot(iris, aes('species', 'sepal_length', fill = 'species')) 
     + geom_boxplot() 
     + theme_minimal()
     + geom_point())
```

```python
(ggplot(iris, aes('species', 'sepal_length')) 
     + geom_boxplot( aes(fill = 'species')) 
     + theme_minimal()
     + geom_point())


#colors specific things
```

```python
bar_iris = iris.groupby("species").agg({"sepal_width": ["mean", "std", "min"]})
bar_iris.columns = ["height_cm", "std", "min"]
bar_iris["species"] = bar_iris.index
bar_iris
```

```python
(ggplot(bar_iris, aes("species", "height_cm", fill = "species" ))
    + geom_bar(stat = "identity"))
```
