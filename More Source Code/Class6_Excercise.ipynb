```python
import warnings
warnings.filterwarnings('ignore')


import pandas as pd
import numpy as np
from plotnine import *
import statsmodels.api as sm
import statsmodels.formula.api as smf


%matplotlib inline
```

```python
# Fill the code in

model = smf.ols(formula = "sugars ~ sodium + fat + protein", data = cereal)
output = model.fit()
output.summary()
```

# Results

```python
#From from the results I am concluding that as the amount 
    # of sugar rises so does sodium and Fat, whereas protein decreases. 
```

```python
from sklearn.preprocessing import StandardScaler #Z-score variables
```

```python
zScore = StandardScaler()
```

```python
# Scale Variables Here
zScore.fit(cereal[["sugars","sodium", "protein", "fat"]]) #only standardize variables that are numeric
Z = zScore.transform(cereal[["sugars","sodium", "protein", "fat"]])
cereal[["sugars","sodium", "protein", "fat"]] = Z
# Run model here
```

```python
cereal.head()
```

```python
model3 = smf.ols(formula = "sugars ~ sodium + protein + fat", data = cereal)
```

```python
output3 = model3.fit()
```

```python
output3.summary()
```

```python
#standard error is alot closer
    #also the y intercept is negative for the zscore model whereas 
    #it is positive for the normal model
```
